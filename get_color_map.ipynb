```python
def rgb_to_hex(l):   
    return '#%02x%02x%02x' % (l[0],l[1],l[2])

def get_color(file,size = 150,k1 = 5):
    im = Image.open(file)
    im.thumbnail((size,size))
    ar = scipy.misc.fromimage(im)
    shape = ar.shape
    ar = ar.reshape(scipy.product(shape[:2]),shape[2])
    ar = ar.astype('float')
    codes, dist = scipy.cluster.vq.kmeans(ar, k1)
    #vecs, dist = scipy.cluster.vq.vq(ar, codes) 
    #counts, bins = scipy.histogram(vecs, len(codes))
    #index_max = scipy.argmax(counts)
    #peak = [int(round(n)) for n in codes[index_max]]
    #color = ''.join(rgb_to_hex(peak))
    return codes
def cell_color(val):
    color = str(val)
    return 'background-color: %s' % color
```

```python
def get_form(folder,k1 = 5,k2 = 10):
    #folder =str(folder)
    filename = './'+ folder + '/*.jpg'
    all_file = glob.glob(filename)
    print (all_file[:1])
    all_color = []
    for file in all_file:
        try:
            color = get_color(file)
        except:
            pass
        if color.shape[1]==3:
            all_color+=(color.tolist())
    ar = np.asarray(all_color)
    codes, dist = scipy.cluster.vq.kmeans(ar, k2)
    code = codes.astype('int')
    colors = []
    
    for i in code:
        color = ''.join(rgb_to_hex(i))
        colors.append(color)
    color_style = []
    for i in colors:
        a = {'background-color':'{}'.format(i)}
        color_style.append(a)
    df = pd.DataFrame(colors,columns = [folder])#+'_color'])
    #df[folder] = ' '
    return df

f = pd.DataFrame()
for folder in ['All','Music','Travel','Technology','Startups','Food','Funny','Gaming','Movies','Fashion','News','Education']:
    form = get_form(folder)
    f = pd.concat([f,form],axis = 1)
f    
```
